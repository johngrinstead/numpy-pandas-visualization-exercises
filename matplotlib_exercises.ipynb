3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
plt.subplot(2,2,1)
x = range(0,50)

y = [math.sqrt(n) for n in x]

plt.plot(x,y)




plt.subplot(2,2,2)
x = range(-10,10)

y = [n ** 3 for n in x]

plt.plot(x,y)




plt.subplot(2,2,3)
x = range(-10,10)

y = [2 ** n for n in x]

plt.plot(x,y)





plt.subplot(2,2,4)
x = range(1,10)

y = [1 / (n + 1) for n in x]

plt.plot(x,y)

plt.show()
```

4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
plt.figure(figsize=(14, 6))


plt.subplot(2,2,1)
x = range(0,50)

y = [math.sqrt(n) for n in x]

plt.plot(x,y, color='red', label='sqrt.x')

plt.legend()


plt.subplot(2,2,2)
x = range(-10,10)

y = [n ** 3 for n in x]

plt.plot(x,y, color='green', label ='x^3')

plt.legend()


plt.subplot(2,2,3)
x = range(-10,10)

y = [2 ** n for n in x]

plt.plot(x,y, color='yellow', label='2^n')


plt.legend()


plt.subplot(2,2,4)
x = range(1,10)

y = [1 / (n + 1) for n in x]

plt.plot(x,y, color='pink', label='1/(x+1)')

plt.legend()

plt.show()
```
