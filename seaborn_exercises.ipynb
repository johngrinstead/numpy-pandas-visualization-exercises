# 4. Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
def get_db_url(user, host, password, db):
    url = (f'mysql+pymysql://{user}:{password}@{host}/{db}')
    return url

from env import host, user, password
```

```python
chipotle_sql = '''
select *
from orders;
'''
```

```python
chipotle_orders = pd.read_sql(chipotle_sql, get_db_url(user, host, password, 'chipotle'))

chipotle_orders.head(10)
```

```python
chipotle_orders['item_price'] = chipotle_orders.item_price.str.replace('$', '').astype(float)

chipotle_orders.head(10)
```

```python
top_four = chipotle_orders.groupby('item_name').quantity.sum().nlargest(4, keep='all')

top_four
```

```python
chipotle_quantity = chipotle_orders.groupby('item_name').quantity.sum()

chipotle_quantity
```

```python

revenue = chipotle_orders.groupby('item_name').item_price.agg('sum')
top_four = revenue.nlargest(4, keep = 'first')
top_four
```

```python
revenue
```

```python
chipotle_quantity_and_revenue = pd.merge(chipotle_quantity, revenue,left_on='item_name',right_on='item_name', how='left')

chipotle_quantity_and_revenue
```

```python
top_four_revenue = chipotle_quantity_and_revenue.sort_values(['quantity', 'item_price'], ascending=[False, False]).head(4)

top_four = top_four_revenue.drop(columns='quantity')
```

```python
top_four.plot.bar()
```
