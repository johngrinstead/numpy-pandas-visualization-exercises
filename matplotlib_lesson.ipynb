# Line Charts
## 1 dependent variable, 1 independent variable

```python
x = list(range(150))

plt.plot(x)
#plot indicates that this is a line chart
##proper syntax is plt.plot(x,y), in the absense of 'y', 'x' will be used for both axis
###if explicitly stated, x and y MUST be the same length 

plt.show()
#show command required to make chart appear 
```

```python
%matplotlib inline
```

```python
plt.plot(range(len(x)), x)
plt.show()
```

```python
plt.plot(x)
plt.xlim(-20, 200)
#place range limits on the x axis

plt.ylim(-5, 160)
#places range limits on y axis

plt.show()
```

```python
plt.plot(x, c='red')
#changes the color of the line to red 

plt.show()
```

```python
from random import randint

# Some random data
x1 = [randint(-5, 25) + n for n in range(150)]
x2 = [x + randint(-6, 6) for x in x1]
#In a line chart you can have multiple lines in one graph


# we can use the alpha kwarg to define how transparent the color is
plt.plot(x1, c='green', alpha=0.6)
plt.plot(x2, c='red', alpha=0.4)
plt.show()
#the graph will not appear until we call plt.show, until then you can add as many lines as you wish
```

# Scatter Charts
## 2 independent variables

```python
# plotting our random data from the last step
plt.scatter(range(len(x1)), x1)
plt.scatter(range(len(x2)), x2)
#plt.scatter indicates it will be a scatter charts
#same as line charts, as many additions can be made to the same chart as desired 

plt.show()
```

```python
plt.scatter(range(len(x1)), x1, s=30, c='red')
plt.scatter(range(len(x2)), x2, s=10, c='green')
#s='number' adjusts the size of each point
#c='color' adjusts the color of the points

plt.show()
```

# Labels, Titles, and Axes

```python
plt.scatter(range(len(x1)), x1, s=30, c='red')
plt.scatter(range(len(x2)), x2, s=10, c='green')
plt.title('A couple of random series')
#takes a string that will be displayed as the title of the table

plt.show()
```

```python
x = list(range(-49, 50))
y = [n ** 2 for n in x]
plt.scatter(x, y)
plt.title('A Quadratic Distribution, $y = x^2$')


plt.xlabel('$x$')
#provides a label for the x axis

plt.ylabel('$x^2$')
#provides a label for the y axis

#using '$' allows the use of LaTeX math

plt.show()
```

```python
x = [1, 2, 3.14, 4]
y = [n ** 2 for n in x]
plt.scatter(x, y)
plt.xticks([1, 2, 3.14, 4], ['one', 'two', '$\pi$', 'four'])
#the second list assigns custom names to the points on the x axis designated by the first list 

plt.show()
```

```python
import math

x = range(-4, 4)
y = [math.sin(n) for n in x]
plt.scatter(x, y)
plt.plot(x, y, ls=':')
# ls=':' assigns the dotted line value to this line chart

plt.title('An approximation of the sine function')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

# Rotating Text and Annotations

```python
x = range(1, 5)
y = x
plt.scatter(x, y)
plt.yticks([1, 2, 3, 4], ['one', 'two', 'three', 'four'], rotation=45)
# rotation='number' assigns how many degrees out of 360 the notches on the y axis will be rotated

plt.xticks(rotation=45)
#assigns rotation for the x axis 

plt.show()
```

```python
import random

weekdays = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'] * 2

x = range(14)
y = [random.randint(1, 10) for _ in x]
plt.plot(x, y)
plt.xticks(x, weekdays, rotation=60)
#rotation can be however much you want

plt.show()
```

```python
x = list(range(-4, 4))
x.append(3.14159)
y = [math.sin(n) for n in x]

plt.scatter(x, y)
plt.plot(x, y, ls=':')

plt.text(0.25, 0, '(0, 0)', fontsize=10, color='blue')
plt.text(1.2, 0.1, '($\pi$, sin($\pi$))', fontsize=14, color='firebrick')
#assigns a point on the graph to include text 


plt.show()
```

```python
x = range(-3, 10)
y = [n * 2 for n in x]
z = [n / 2 for n in x]

plt.plot(x, y, c='orange')
plt.plot(x, z, c='green')

plt.annotate('Intersection', xy=(0, 0), xytext=(-3, 5),
             arrowprops={'facecolor': 'blue'})
#used to create an arrow to draw attention to a point on chart

plt.show()
```

# Figure Size and Legends

```python
x = range(-4, 5)
sin_x = [math.sin(n) for n in x]
cos_x = [math.cos(n) for n in x]

plt.figure(figsize=(14, 6)) # (width, height)
#specifies dimensions of entire figure

plt.scatter(x, sin_x, c='blue')
plt.scatter(x, cos_x, c='orange')

plt.plot(x, sin_x, c='blue', label='$\sin(x)$')
plt.plot(x, cos_x, c='orange', label='$\cos(x)$')
#label='string' designates a name in the legend for the graph

plt.legend(loc='upper right')
#plt.legend REQUIRED for a legend to appear at all

plt.title('Approximate sin and cosine curves')
plt.xlabel('$x$')

plt.show()
```

# Subplots

```python
n_rows = 1
n_cols = 2

# some data to play with
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]
z = [1, 2, 3, 4, 5]

# plot the first subplot
plt.subplot(n_rows, n_cols, 1)
#to plot a subplot you need to designate which row, which coloumn, and which order a specific graph is in

plt.plot(x, y)
plt.title('x ~ y')

# the second subplot
plt.subplot(n_rows, n_cols, 2)
plt.plot(x, z)
plt.title('x ~ z')

plt.show()
```

```python
# some data to play with
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]
z = [1, 2, 3, 4, 5]

# plot the first subplot
plt.subplot(121)
#designation can be shorthand instead of (row, column, order)
plt.plot(x, y)
plt.title('x ~ y')

# the second subplot
plt.subplot(122)
plt.plot(x, z)
plt.title('x ~ z')

plt.show()
```

```python
# some data to play with
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]
z = [1, 2, 3, 4, 5]

# plot the first subplot
plt.subplot(121)
plt.plot(x, y)
plt.title('x ~ y')

# the second subplot
plt.subplot(122)
plt.plot(x, z)
plt.title('x ~ z')

plt.suptitle('Subplots Demo')
#provides a title for the ENTIRE figure 

plt.show()
```

# Histograms

```python
x = [1, 1, 2, 3, 3, 3, 4, 4, 5]

plt.hist(x)
#plt.hist() indicates the graph will be a histogram

plt.show()
```

```python
x = [1, 1, 2, 3, 3, 3, 4, 4, 5]

plt.hist(x, bins=[0, 1, 2, 3, 4, 5, 6])
#bins=[list] manually assigns bins for data to be sorted in, otherwise the computer will automatically assume the best bin arrangement

plt.show()
```

```python
x = [1, 1, 2, 3, 3, 3, 4, 4, 5]

plt.figure(figsize=(16, 6))
plt.suptitle(f'x = {x}')

plt.subplot(131)
plt.hist(x, bins=[0, 1, 2, 3, 4, 5, 6])
plt.title('align="middle" (default)')
#align='middle','left','right' will tell the computer how to format the bar compared to the tick on the y axis

plt.subplot(132)
plt.hist(x, bins=[0, 1, 2, 3, 4, 5, 6], align='left')
plt.title('align="left"')

plt.subplot(133)
plt.hist(x, bins=[0, 1, 2, 3, 4, 5, 6], align='right')
plt.title('align="right"')

plt.show()
```

```python
x1 = [randint(1, 5) for _ in range(20)]
x2 = [randint(1, 5) for _ in range(20)]
#multiple histograms can be placed on top of each other

plt.hist(x1, bins=[0, 1, 2, 3, 4, 5, 6], align='left', edgecolor='black', alpha=0.5, color='blue')
plt.hist(x2, bins=[0, 1, 2, 3, 4, 5, 6], align='left', edgecolor='black', alpha=0.5, color='blue')
#edgecolor='color' assigns a color to the border of the bar
#alpha=0-1 assigns how transparent the bar is on a scale of 0(invisible) to 1(solid)

plt.show()
```

```python
x1 = [randint(1, 5) for _ in range(20)]
x2 = [randint(1, 5) for _ in range(20)]

plt.hist(x1, bins=[1, 2, 3, 4, 5, 6], align='left', edgecolor='black', alpha=0.4, label='$x_1$')
plt.hist(x2, bins=[1, 2, 3, 4, 5, 6], align='left', edgecolor='black', alpha=0.4, label='$x_2$')
plt.legend()

plt.show()
```

```python
x1 = [randint(1, 5) for _ in range(20)]
x2 = [randint(1, 5) for _ in range(20)]

plt.hist((x1, x2), bins=[1, 2, 3, 4, 5, 6], align='left', edgecolor='black', alpha=0.6, label=['$x_1$', '$x_2$'])
#listing the data as a tuple makes the bars appear side by side as opposed to overlapping 

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
for i in range(1, 5):
    n = 10 ** i
    x1 = [randint(1, 6) for _ in range(n)]
    x2 = [randint(1, 6) for _ in range(n)]

    total = [x + y for x, y in zip(x1, x2)]

    plt.subplot(2, 2, i)
    plt.hist(total, bins=range(2, 14), align='left', color='white', edgecolor='black')
    plt.title(f'n = {n}')

plt.suptitle('Sum of 2 dice rolls')
plt.show()
```
